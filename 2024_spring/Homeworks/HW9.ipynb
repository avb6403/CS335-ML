```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles, make_moons
from mlxtend.plotting import plot_decision_regions
```

```python
def load_planar_dataset():
    np.random.seed(1)
    m = 400 # number of examples
    N = int(m/2) # number of points per class
    D = 2 # dimensionality
    X = np.zeros((m,D)) # data matrix where each row is a single example
    Y = np.zeros(m, dtype='uint8') # labels vector (0 for red, 1 for blue)
    a = 4 # maximum ray of the flower

    for j in range(2):
        ix = range(N*j,N*(j+1))
        t = np.linspace(j*3.12,(j+1)*3.12,N) + np.random.randn(N)*0.2 # theta
        r = a*np.sin(4*t) + np.random.randn(N)*0.2 # radius. 4 corresponds to the number of petals
        X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
        Y[ix] = j

    return X, Y
```

Below is the model from the class: $x\to h\to a=\sigma(h)\to o\to Loss(o,y)$, where $Loss$ is MSE:
$$
Loss(o,y) = \frac{1}{N} \sum_{i=1}^{N}\left( o^{(i)}-y^{(i)}\right)^2.
$$

For the binary classification we use binary cross-entropy loss:
$$
Loss(\sigma(o^{(i)}),y^{(i)}) = -\frac{1}{N}\sum_{i=1}^{N}\left( y^{(i)}\log \sigma(o^{(i)}) + (1-y^{(i)})\log\left( 1-\sigma(o^{(i)})\right) \right)
$$

```python
class Loss():
  def __init__(self):
    self.x_in = None
    self.y_in = None
    self.eps = 10**(-6)

  def forward(self, x_in, y_in):
    self.x_in = x_in
    self.y_in = y_in
    # return np.sum((self.x_in-self.y_in)**2)/len(self.x_in) # Initial
    return -(np.sum(self.y_in * np.log(self.x_in + self.eps) + (1 - self.y_in) * np.log(1 - self.x_in + self.eps)))/len(self.x_in) # Q2

  def backward(self):
    # return 2*(self.x_in-self.y_in)/len(self.x_in) # Initial
    return -(self.y_in/(self.x_in + self.eps))+((1-self.y_in)/(1 - self.x_in + self.eps))/len(self.x_in) # Q2
```

```python
class Activation():
  def __init__(self):
    self.x_in = None

  def forward(self, x_in):
    self.x_in = x_in
    return 1/(1+np.exp(-self.x_in))

  def backward(self, grad, lr=1e-3):
    return grad* np.exp(self.x_in)/(1+np.exp(self.x_in))**2
```

```python
class Linear():
  def __init__(self, input_size, output_size):
    self.W = np.random.random((output_size, input_size))*0.01
    self.b = np.random.random((output_size, 1))*0.01

    self.grad_W = np.zeros(self.W.shape)
    self.grad_b = np.zeros(self.b.shape)

    self.x_in = None

  def forward(self, x_in):
    self.x_in = x_in
    return (self.W.dot(self.x_in.T) + self.b).T

  def backward(self, grad, lr=1e-3):
    self.grad_W = grad.T @ self.x_in
    self.grad_b = grad.T @ np.ones((len(self.x_in),1))
    grad = grad @ self.W
    self.W -= lr * self.grad_W
    self.b -= lr * self.grad_b
    return grad
```

```python
class MyNN():
  def __init__(self, input_size=1, hidden_size=3, output_size=1) -> None:
    self.input_size = input_size
    self.hidden_size = hidden_size
    self.output_size = output_size
    self.model = [Linear(self.input_size,self.hidden_size), Activation(), Linear(self.hidden_size,self.output_size), Activation()] # Add activation as a last layer
    self.loss = Loss()

  def fit(self, x, y, lr=0.01, n_epoch = 1000, warm_start=True):
    self.loss_array = []
    self.lr = lr
    self.n_epoch = n_epoch
    if not warm_start:
      self.model = [Linear(self.input_size,self.hidden_size), Activation(), Linear(self.hidden_size,self.output_size), Activation()] # Add activation as a last layer
      self.loss = Loss()
    for epoc in range(1,self.n_epoch):
      x_in = x
      y_in = y
      for m in self.model:
        x_in = m.forward(x_in)
      self.loss_array.append(self.loss.forward(x_in, y_in))
      grad = self.loss.backward()
      for i in range(len(self.model)-1, -1, -1):
        grad = self.model[i].backward(grad, lr=self.lr)

  def predict(self, x):
      x_in = x.reshape(len(x),-1)
      for m in self.model:
        x_in = m.forward(x_in)
      # return np.squeeze(x_in) # Initial
      predictions = (x_in > 0.5).astype(int)
      return np.squeeze(predictions) # Q3
```

The behavior of the loss function during training provides insights into the learning process and the effectiveness of chosen hyperparameters, such as the learning rate. Here are my observations:

Jumps: If large jumps in the loss function are observed, it often indicates that the learning rate is too high. In this case, the updates to the weights and biases are too large, causing the model to overshoot the minimum of the loss function and potentially diverge.

Drops: A steady drop in the loss function is a good sign, indicating that the model is learning and converging towards a minimum. However, if the drop is too rapid, it might indicate that the learning rate is too high, and the model might miss the global minimum.

Oscillations: If the loss function oscillates, it means that the model is bouncing around the minimum without settling down. This can be caused by a learning rate that is slightly too high. Reducing the learning rate can help the model converge more smoothly.

Plateaus: A plateau in the loss function might indicate that the model has reached a local minimum or that the learning rate is too low. In this case, increasing the learning rate or using techniques like learning rate scheduling can help the model escape the plateau and continue learning.

Regarding the convexity of the loss function with respect to the weights W and biases b, it depends on the architecture of the neural network and the choice of the loss function:

* For a simple linear regression model with a mean squared error loss function, the loss function is convex with respect to W and b, and there is a single global minimum.
* For more complex models, such as neural networks with non-linear activation functions, the loss function is generally non-convex. This means that there can be multiple local minima, and the optimization algorithm might not converge to the global minimum.
